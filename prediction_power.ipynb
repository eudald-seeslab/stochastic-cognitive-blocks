```python
import pandas as pd
from mmsbm import MMSBM
```

```python
sampling = 6
# Number of groups of users
k = 3
# Number of groups of items
l = 3
# Iterations
# To plateau the coefficients the minimum is 500
iterations = 500

runs = 20
```

```python
def predict_missing_data(train, test, iteration):

    mmsbm = MMSBM(
        user_groups=k,
        item_groups=l,
        iterations=iterations,
        sampling=sampling,
        seed=1714
    )
    accuracies = mmsbm.cv_fit(train, folds=5)

    pred_matrix = mmsbm.predict(test)
    results = mmsbm.score()

    return iteration, results["stats"]["accuracy"], results["stats"]["one_off_accuracy"]
```

# 40% missing data

```python
res40 = dict()
for i in range(runs):
    train = pd.read_csv(f"data/missings/missing_0.4_train_{i + 1}.csv")
    train.value = train.value.astype("Int64")
    train.value = train.value - 1
    answers = pd.read_csv(f"data/missings/missing_0.4_answers_{i + 1}.csv")
    answers.value = answers.value - 1

    res40[i] = predict_missing_data(train, answers, i)
```

```python
pd.DataFrame.from_dict(res40, orient="index").rename(columns={0: "grade", 1: "mean", 2: "sd"}).to_csv("output/missing_pred_40.csv")
```
